# Sales Analysis Notebook
Exploratory analysis of regional sales data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Generate synthetic data

```python
np.random.seed(42)
n = 200
dates = pd.date_range('2023-01-01', periods=n, freq='D')

df_north = pd.DataFrame({
    'date': dates,
    'revenue': np.random.normal(5000, 800, n) + np.linspace(0, 1500, n),
    'units': np.random.poisson(120, n),
    'returns': np.random.poisson(8, n),
    'region': 'North'
})

df_south = pd.DataFrame({
    'date': dates,
    'revenue': np.random.normal(4200, 600, n) + np.linspace(0, 800, n),
    'units': np.random.poisson(95, n),
    'returns': np.random.poisson(12, n),
    'region': 'South'
})

df_west = pd.DataFrame({
    'date': dates,
    'revenue': np.random.normal(6100, 1100, n) + np.linspace(0, 2200, n),
    'units': np.random.poisson(150, n),
    'returns': np.random.poisson(6, n),
    'region': 'West'
})
```

---
## Section 1 — North Region

```python
# Basic stats - North
print('North region summary')
print(df_north[['revenue', 'units', 'returns']].describe())
```

```python
# Compute metrics - North
north_avg_revenue = df_north['revenue'].mean()
north_total_revenue = df_north['revenue'].sum()
north_return_rate = df_north['returns'].sum() / df_north['units'].sum()
north_revenue_per_unit = df_north['revenue'].sum() / df_north['units'].sum()

print(f'Avg daily revenue: {north_avg_revenue:.2f}')
print(f'Total revenue: {north_total_revenue:.2f}')
print(f'Return rate: {north_return_rate:.4f}')
print(f'Revenue per unit: {north_revenue_per_unit:.2f}')
```

```python
# 7-day rolling average revenue - North
df_north['rolling_revenue'] = df_north['revenue'].rolling(7).mean()
```

```python
# Plot revenue trend - North
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(df_north['date'], df_north['revenue'], alpha=0.4, label='Daily')
ax.plot(df_north['date'], df_north['rolling_revenue'], label='7d avg', linewidth=2)
ax.set_title('North — Revenue Trend')
ax.set_xlabel('Date')
ax.set_ylabel('Revenue')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# Monthly aggregation - North
north_monthly = df_north.copy()
north_monthly['month'] = north_monthly['date'].dt.to_period('M')
north_monthly = north_monthly.groupby('month').agg(
    revenue=('revenue', 'sum'),
    units=('units', 'sum'),
    returns=('returns', 'sum')
).reset_index()
north_monthly['return_rate'] = north_monthly['returns'] / north_monthly['units']
print(north_monthly)
```

```python
# Plot monthly bar - North
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar(north_monthly['month'].astype(str), north_monthly['revenue'])
ax.set_title('North — Monthly Revenue')
ax.set_xlabel('Month')
ax.set_ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Distribution of daily units - North
fig, ax = plt.subplots(figsize=(6, 4))
ax.hist(df_north['units'], bins=20, edgecolor='black')
ax.set_title('North — Units Sold Distribution')
ax.set_xlabel('Units')
ax.set_ylabel('Frequency')
plt.tight_layout()
plt.show()
```
